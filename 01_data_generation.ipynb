```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
random.seed(42)
np.random.seed(42)

fake_us = Faker('en_US')
fake_uk = Faker('en_GB')
```

```python
random.seed(42)
np.random.seed(42)

fake_us = Faker('en_US')
fake_uk = Faker('en_GB')
```

```python
departments = ['Flight Reservations', 'Hotel Booking', 'Visa Processing', 
               'Customer Support', 'Travel Insurance', 'Cruise Booking',
               'Holiday Packages', 'Corporate Travel Desk', 'Refunds & Cancellations']

processes = ['Inbound Voice', 'Outbound Voice', 'Chat Support', 'Email Support', 'Blended']

shifts = ['Day Shift', 'Night Shift (US)', 'Night Shift (UK)', 'Rotational']

cities_india_bpo_hubs = ['Pune', 'Bangalore', 'Gurgaon', 'Noida', 'Hyderabad', 
                          'Mumbai', 'Chennai', 'Kolkata']
```

```python
data = []

for i in range(10000):
    country_process = random.choice(['USA', 'UK'])
    fake = fake_us if country_process == 'USA' else fake_uk
    
    record = {
        'Employee_ID': f'BPO{10001+i}',
        'Name': fake.name(),
        'Age': random.randint(21, 48),
        'Gender': random.choice(['Male', 'Female']),
        'City': random.choice(cities_india_bpo_hubs),
        'Process_Country': country_process,
        'Department': random.choice(departments),
        'Process_Type': random.choice(processes),
        'Shift': random.choice(shifts),
    }
    data.append(record)

print(f"Generated {len(data)} records!")
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
df.to_csv('travel_bpo_employee_data_simplified.csv', index=False)
print("CSV file saved successfully!")
```
