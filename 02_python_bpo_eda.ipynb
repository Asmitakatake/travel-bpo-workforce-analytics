```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('travel_bpo_employee_data_simplified.csv')
print(f"Total Rows: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
df.head(10)
```

```python
print("=== DATA TYPES ===")
print(df.info())
print("\n=== BASIC STATISTICS ===")
print(df.describe())
```

```python
print("=== MISSING VALUES ===")
print(df.isnull().sum())
print(f"\nTotal Missing Values: {df.isnull().sum().sum()}")
print(f"\nDuplicate Rows: {df.duplicated().sum()}")
```

```python
plt.figure(figsize=(6, 4))
gender_counts = df['Gender'].value_counts()
colors = ['#185FA5', '#E24B4A']
plt.pie(gender_counts.values, 
        labels=gender_counts.index, 
        autopct='%1.1f%%', 
        colors=colors,
        startangle=90)
plt.title('Gender Distribution of BPO Employees', 
          fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('chart1_gender_distribution.png', dpi=150)
plt.show()
print("Chart 1 saved!")
```

```python
plt.figure(figsize=(12, 5))
dept_count = df['Department'].value_counts()
colors = sns.color_palette('Blues_d', len(dept_count))
bars = plt.bar(dept_count.index, dept_count.values, color=colors)
plt.title('Employee Headcount by Department', 
          fontsize=14, fontweight='bold')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
for bar, val in zip(bars, dept_count.values):
    plt.text(bar.get_x() + bar.get_width()/2, 
             bar.get_height() + 5, 
             str(val), ha='center', fontsize=9)
plt.tight_layout()
plt.savefig('chart2_department_headcount.png', dpi=150)
plt.show()
print("Chart 2 saved!")
```

```python
plt.figure(figsize=(10, 5))
city_count = df['City'].value_counts()
colors = sns.color_palette('Greens_d', len(city_count))
bars = plt.bar(city_count.index, city_count.values, color=colors)
plt.title('Employee Headcount by City', 
          fontsize=14, fontweight='bold')
plt.xlabel('City')
plt.ylabel('Number of Employees')
for bar, val in zip(bars, city_count.values):
    plt.text(bar.get_x() + bar.get_width()/2, 
             bar.get_height() + 5, 
             str(val), ha='center', fontsize=9)
plt.tight_layout()
plt.savefig('chart3_city_headcount.png', dpi=150)
plt.show()
print("Chart 3 saved!")
```

```python
plt.figure(figsize=(8, 5))
shift_count = df['Shift'].value_counts()
colors = ['#1D9E75', '#185FA5', '#BA7517', '#E24B4A']
bars = plt.bar(shift_count.index, shift_count.values, color=colors)
plt.title('Employee Distribution by Shift', 
          fontsize=14, fontweight='bold')
plt.xlabel('Shift Type')
plt.ylabel('Number of Employees')
plt.xticks(rotation=15)
for bar, val in zip(bars, shift_count.values):
    plt.text(bar.get_x() + bar.get_width()/2, 
             bar.get_height() + 5, 
             str(val), ha='center', fontsize=10)
plt.tight_layout()
plt.savefig('chart4_shift_distribution.png', dpi=150)
plt.show()
print("Chart 4 saved!")
```

```python
plt.figure(figsize=(6, 4))
country_count = df['Process_Country'].value_counts()
colors = ['#185FA5', '#E24B4A']
bars = plt.bar(country_count.index, 
               country_count.values, 
               color=colors, width=0.4)
plt.title('US vs UK Process Country Split', 
          fontsize=14, fontweight='bold')
plt.xlabel('Process Country')
plt.ylabel('Number of Employees')
for bar, val in zip(bars, country_count.values):
    plt.text(bar.get_x() + bar.get_width()/2, 
             bar.get_height() + 10, 
             str(val), ha='center', fontsize=11, fontweight='bold')
plt.tight_layout()
plt.savefig('chart5_country_split.png', dpi=150)
plt.show()
print("Chart 5 saved!")
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Age'], bins=20, color='#185FA5', 
         edgecolor='white', linewidth=0.5)
plt.title('Age Distribution of BPO Employees', 
          fontsize=14, fontweight='bold')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.axvline(df['Age'].mean(), color='red', 
            linestyle='--', linewidth=2, 
            label=f"Mean Age: {df['Age'].mean():.1f}")
plt.legend()
plt.tight_layout()
plt.savefig('chart6_age_distribution.png', dpi=150)
plt.show()
print(f"Chart 6 saved! Average Age: {df['Age'].mean():.1f}")
```
